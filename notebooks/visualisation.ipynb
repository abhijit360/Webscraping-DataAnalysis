```python
import numpy , pandas
import matplotlib.pyplot as plt
```

```python
data_frame = pandas.read_csv("../athleteDATA.csv")
data_frame.head(19)
```

```python
# hundred_meter_mask = data_frame['event_name'] == "100m"
hundred_meter_data = data_frame[data_frame['event_name'] == "100m"]

finals_data = hundred_meter_data[hundred_meter_data["Round"] == "Finals"]

# for round in hundred_meter_data["Round"]:
#     round_mask = hundred_meter_data['Round'] == round
#     round_data = hundred_meter_data[round_mask]
#     print(round_data["Round"])
#     print("next Round")
```

```python
mens_data = finals_data[finals_data['gender']=="Men"]
womens_data = finals_data[finals_data['gender']=="Women"]


# print(mens_data["Time"])
positions = [1,2]  # x-axis positions for the box plots
data_to_plot = [pandas.to_numeric(mens_data["Time"], errors="coerce"), pandas.to_numeric(womens_data["Time"], errors="coerce")]

fig = plt.figure()
# plt.boxplot(data_to_plot, positions=positions, labels=["Men","Women"])
plt.boxplot(x=data_to_plot[0], labels=['men'])
plt.xlabel("Gender")
plt.ylabel("Time(seconds)")
plt.yticks
plt.title("Time distribution of events by Gender")
fig.show()
```
